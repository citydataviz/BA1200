```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(2004)
```

## Histogram

```python
continuous_data = np.random.normal(loc=50, scale=10, size=1000)
plt.hist(continuous_data, bins=10, color='lightgreen')
plt.title('Histogram')
plt.ylabel('Frequency')
plt.xlabel('Value')

plt.show()
```

## Bar Plot

```python
category_names = ['Category A', 'Category B', 'Category C', 'Category D']
category_values = [23, 45, 56, 78]

plt.bar(category_names, category_values, color='skyblue')
plt.title('Bar Chart')
plt.ylabel('Values')
plt.xlabel('Categories')
plt.show()
```
