### Add a legend (mean, var, std)

```python
df.x123.mean()
df.x123.var()
df.x123.std()
```

```python

# Function to calculate mean, variance, and standard error of "y" 
def calc_stats(data):
    mean = data.mean()
    variance = data.var()
    std_error = data.std()
    return mean, variance, std_error

fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (7, 7), 
                       sharex = 'col', sharey = 'row')

def poly(x,y,subplot) :
    mean, variance, std_error = calc_stats(y) 
    slope, intercept = np.polyfit(x, y, 1)
    subplot.scatter(x,y, s=10)
    subplot.plot(x,slope*x + intercept, color='red')
    subplot.legend([f'Mean: {mean:.2f}\nVariance: {variance:.2f}\nStd Error: {std_error:.2f}'])

poly(df.x123,df.y1,ax[0][0])
poly(df.x123,df.y2,ax[0][1])
poly(df.x123,df.y3,ax[1][0])
poly(df.x4,df.y4,ax[1][1])

plt.tight_layout()
plt.show()

```
